# QDLDL vs NumPy/SciPy

This notebook solves a symmetric positive definite linear system with the local `solvers.qdldl` module, NumPy, and SciPy sparse linear algebra.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / "CMakeLists.txt").exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "build"))

import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from solvers import qdldl
```

```python
rng = np.random.default_rng(7)
n = 8
M = rng.normal(size=(n, n))
A = M.T @ M + 0.5 * np.eye(n)
b = rng.normal(size=n)

x_qdldl = qdldl.solve(A, b)["x"]
x_numpy = np.linalg.solve(A, b)
x_scipy = spla.spsolve(sp.csc_matrix(A), b)

print("QDLDL x:", x_qdldl)
print("NumPy max error:", np.max(np.abs(x_qdldl - x_numpy)))
print("SciPy max error:", np.max(np.abs(x_qdldl - x_scipy)))
print("Residual inf norm:", np.linalg.norm(A @ x_qdldl - b, ord=np.inf))
```

```python
assert np.allclose(x_qdldl, x_numpy, rtol=1e-8, atol=1e-10)
assert np.allclose(x_qdldl, x_scipy, rtol=1e-8, atol=1e-10)
```
